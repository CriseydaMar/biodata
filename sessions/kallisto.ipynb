```bash
# This needs to be run from the file to preserve output

wt="H7 H8 H9"
fasta_dir=../fastq/

for sample in $wt
do
    echo "Running kallisto quant for sample "$sample
    output=../kallisto/$sample
    if [ ! -d "$output" ]; then
        mkdir -p $output
    fi
    read1=$fasta_dir$(ls $fasta_dir | grep $sample | grep R1)
    read2=$fasta_dir$(ls $fasta_dir | grep $sample | grep R2)
    kallisto quant -i ../kallisto/cele_tx -o $output -b 100 <(zcat $read1) <(zcat $read2)
done
```
